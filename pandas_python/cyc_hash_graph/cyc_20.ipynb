```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# read json file
cc20_df = pd.read_json('cc_20.json')
cc22_df = pd.read_json('cc_22.json')
cc24_df = pd.read_json('cc_24.json')
cc26_df = pd.read_json('cc_26.json')
cc27_df = pd.read_json('cc_27.json')

# set size
plt.figure(figsize=(10, 6))

# cc_20 plot
plt.plot(cc20_df['partition'], cc20_df['load imbalance'], '^-', label='1.04M', color='#40BE73')
# cc_22 plot
plt.plot(cc22_df['partition'], cc22_df['load imbalance'], 'D-', label='4.19M', color='#ff6666')
# cc_24 plot
plt.plot(cc24_df['partition'], cc24_df['load imbalance'], 's-', label='16.77M', color='orange')
# cc_26 plot
plt.plot(cc26_df['partition'], cc26_df['load imbalance'], 'o-', label='67.10M', color='#48B2F9')
# cc_26 plot
plt.plot(cc27_df['partition'], cc27_df['load imbalance'], 'x-', label='134.21M', color='#736EF5')


# title/label
plt.title('Cyclic Hash - Kronecker graph')
plt.xlabel('Number of Partitions')
plt.ylabel('Load Imbalance')
# log scale
plt.xscale('log')  
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```
