```python
import pandas as pd
import matplotlib.pyplot as plt

# read json file
snake_df = pd.read_json('snake_24outfile.json')
cyc_df = pd.read_json('cyc_24outfile.json')
range_df = pd.read_json('range_24outfile.json')
jch_df = pd.read_json('jch_24outfile.json')
rotation_df = pd.read_json('rotation_24outfile.json')
# set size
plt.figure(figsize=(10, 6))
# cyc_data plot
plt.plot(cyc_df['partition'], cyc_df['load imbalance'], 'D-', label='cyc', color='#FF3333')
# jch data plot
plt.plot(jch_df['partition'], jch_df['load imbalance'], 's-', label='jch', color= '#A2E15A')
# range_data plot
plt.plot(range_df['partition'], range_df['load imbalance'], '^-', label='range', color='orange')
# rotation plot
plt.plot(rotation_df['partition'], rotation_df['load imbalance'], 'x-', label='rotation', color='gray')
# snake plot
plt.plot(snake_df['partition'], snake_df['load imbalance'], 'o-', label ='snake', color='#4D94FF')
# title/labels and other plot tweaks
plt.title('Road.sg graph: Load Imbalance vs Partition')
plt.xlabel('Number of Partitions')
plt.ylabel('Load Imbalance')
# log scale
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```
