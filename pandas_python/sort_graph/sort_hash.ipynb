```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# read json file
sort20_df = pd.read_json('sort_20.json')
sort22_df = pd.read_json('sort_22.json')
sort24_df = pd.read_json('sort_24.json')
sort26_df = pd.read_json('sort_26.json')
sort27_df = pd.read_json('sort_27.json')

# set size
plt.figure(figsize=(10, 6))

# sort_20 plot
plt.plot(sort20_df['partition'], sort20_df['load imbalance'], '^-', label='1.04M', color='#40BE73')
# sort_22 plot
plt.plot(sort22_df['partition'], sort22_df['load imbalance'], 'x-', label='4.19M', color='#736EF5')
# sort_24 plot
plt.plot(sort24_df['partition'], sort24_df['load imbalance'], 's-', label='16.77M', color='orange')
# sort_26 plot
plt.plot(sort26_df['partition'], sort26_df['load imbalance'], 'o-', label='67.10M', color='#48B2F9')
# sort_27 plot
plt.plot(sort27_df['partition'], sort27_df['load imbalance'], 'D-', label='134.21M', color='#ff6666')

# title/label
plt.title('Sort Hash - Unshuffle Kronecker graph 2^20 - 2^27')
plt.xlabel('Number of Partitions')
plt.ylabel('Load Imbalance')
# log scale
plt.xscale('log')  
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# read json file
cc_27_df = pd.read_json('cc_27.json')
jch_27_df = pd.read_json('jch_27.json')
rot_27_df = pd.read_json('rot_27.json')
snake_27_df = pd.read_json('snake_27.json')
sort27_df = pd.read_json('sort_27.json')

# set size
plt.figure(figsize=(10, 6))

# cc_27 plot
plt.plot(cc_27_df['partition'], cc_27_df['load imbalance'], '^-', label='cyclic', color='#40BE73')
# jch_27 plot
plt.plot(jch_27_df['partition'], jch_27_df['load imbalance'], 'x-', label='jch', color='#736EF5')
# rot_27 plot
plt.plot(rot_27_df['partition'], rot_27_df['load imbalance'], 's-', label='rotation', color='orange')
# snake_27 plot
plt.plot(snake_27_df['partition'], snake_27_df['load imbalance'], 'o-', label='snake', color='#48B2F9')
# sort_27 plot
plt.plot(sort27_df['partition'], sort27_df['load imbalance'], 'D-', label='sort', color='#ff6666')

# title/label
plt.title('Load Imbalance - Unshuffle Kronecker graph 2^27')
plt.xlabel('Number of Partitions')
plt.ylabel('Load Imbalance')
# log scale
plt.xscale('log')  
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```
