```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# read json file
jch20_df = pd.read_json('jchS_20.json')
jch22_df = pd.read_json('jchS_22.json')
jch24_df = pd.read_json('jchS_24.json')
jch26_df = pd.read_json('jchS_26.json')
jch27_df = pd.read_json('jchS_27.json')

# set size
plt.figure(figsize=(10, 6))

# jch_20 plot
plt.plot(jch20_df['partition'], jch20_df['load imbalance'], '^-', label='1.04M', color='#40BE73')
# jch_22 plot
plt.plot(jch22_df['partition'], jch22_df['load imbalance'], 'x-', label='4.19M', color='#736EF5')
# jch_24 plot
plt.plot(jch24_df['partition'], jch24_df['load imbalance'], 's-', label='16.77M', color='orange')
# jch_26 plot
plt.plot(jch26_df['partition'], jch26_df['load imbalance'], 'o-', label='67.10M', color='#48B2F9')
# jch_27 plot
plt.plot(jch27_df['partition'], jch27_df['load imbalance'], 'D-', label='134.21M', color='#ff6666')

# title/label
plt.title('JCH Hash - Shuffle Kronecker graph')
plt.xlabel('Number of Partitions')
plt.ylabel('Load Imbalance')
# log scale
plt.xscale('log')  
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# read json file
snake_df = pd.read_json('uni_snake_outfile.json')
cyc_df = pd.read_json('uni_cyc_outfile.json')
range_df = pd.read_json('uni_range_outfile.json')
jch_df = pd.read_json('uni_jch_outfile.json')
rotation_df = pd.read_json('uni_rotation_outfile.json')

# set size
plt.figure(figsize=(12, 6))

# cyc_data plot
plt.plot(cyc_df['partition'], cyc_df['load imbalance'], '^-', label='cyc', color='blue')
# range_data plot
plt.plot(range_df['partition'], range_df['load imbalance'], 's-', label='range', color='red')
# jch data plot
plt.plot(jch_df['partition'], jch_df['load imbalance'], 'o-', label='jch', color='green')
# rotation plot
plt.plot(rotation_df['partition'], rotation_df['load imbalance'], 'D-', label='rotation', color='purple')
# snake plot
plt.plot(snake_df['partition'], snake_df['load imbalance'], 'x-', label ='snake', color='orange')

# title/label
plt.title('Uniform graph, scale: 67M, 34K partitions')
plt.xlabel('Partition')
plt.ylabel('Load Imbalance')
plt.xscale('log')  
plt.legend()
plt.grid(True)
plt.tight_layout()
# show
plt.show()
```
